```python
from diffusers import StableDiffusionPipeline
import gradio as gr

# Load the Stable Diffusion model
model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
model.to("cuda")  # This will use the GPU

# Function to generate digital art with customized parameters
def generate_digital_art(prompt, guidance_scale=7.5, num_inference_steps=50):
    """
    Generate a digital artwork based on the prompt with customization options:
    - guidance_scale: Controls the creativity (higher values make the image more aligned with the prompt)
    - num_inference_steps: Controls quality and detail (higher values improve quality but increase time)
    """

    # Define negative prompts to avoid overly realistic elements while allowing inspiration
    negative_prompt = "photorealistic, photography, real life, realism"

    # Artistic styles to include in the prompt
    artistic_styles = "digital art, concept art, illustration, vibrant colors, painterly style"

    # Combine the prompt with artistic styles
    combined_prompt = f"{prompt}, {artistic_styles}"

    # Generate the image
    image = model(
        prompt=combined_prompt,
        negative_prompt=negative_prompt,
        guidance_scale=guidance_scale,
        num_inference_steps=num_inference_steps
    ).images[0]

    return image

# Set up Gradio Interface with input parameters
interface = gr.Interface(
    fn=generate_digital_art,
    inputs=[
        gr.Textbox(label="Art Prompt", placeholder="Describe the art you want"),
        gr.Slider(label="Guidance Scale (Creativity)", minimum=5, maximum=20, step=0.5, value=7.5),
        gr.Slider(label="Inference Steps (Quality)", minimum=10, maximum=100, step=5, value=50)
    ],
    outputs="image",
    title="Digital Art Generator",
    description="Create digital art inspired by real life with custom settings."
)

interface.launch()
```
